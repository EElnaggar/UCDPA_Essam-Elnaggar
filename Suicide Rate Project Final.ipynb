```python
# Import the GDP table by country for the last year

table_gdp = pd.read_html('https://www.macrotrends.net/countries/ranking/gdp-per-capita')
df_gdp = table_gdp[0]
df_gdp.head(5)
```

```python
# Drop all columns except the 2018 colum

df_gdp.drop(['2021', '2020', '2019', '2017'], inplace=True, axis=1)
```

```python
df_gdp.head(5)
```

```python
df_gdp.rename(columns={"Country Name":"country"}, inplace=True)
```

```python
df_gdp.count()
```

```python
# Split string column into two new columns

df_gdp[['currency', 'gdp_per_capita']] = df_gdp['2018'].str.split("$", expand = True)
print(df_gdp)
```

```python
df_gdp.drop(['2018', 'currency'], inplace=True, axis = 1)
```

```python
df_gdp.head(5)
```

```python
# Removing a comma from a column values in Pandas DataFrame

df_gdp['gdp_per_capita'] = df_gdp['gdp_per_capita'].str.replace(',', '')
```

```python
df_gdp.head(5)
```

```python
df_gdp.dtypes
```

```python
df_gdp.info()
```

```python
# convert "gdp_2018" from object to int64

df_gdp = df_gdp.astype({'gdp_per_capita':'int'})
```

```python
df_gdp.info()
```

```python
# create a list of our conditions
conditions = [
    (df_gdp['gdp_per_capita'] <= 1026),
    (df_gdp['gdp_per_capita'] > 1026) & (df_gdp['gdp_per_capita'] <= 3995),
    (df_gdp['gdp_per_capita'] > 3995) & (df_gdp['gdp_per_capita'] <= 12375),
    (df_gdp['gdp_per_capita'] > 12375)
    ]

# create a list of the values we want to assign for each condition
values = ['Low-income', 'Lower-middle-income', 'Upper-middle-income', 'High-income']

# create a new column and use np.select to assign values to it using our lists as arguments
df_gdp['income_group'] = np.select(conditions, values)

# display updated DataFrame
df_gdp.head()
```

```python
# new DataFrame based on country and suicides_per_100k_pop agg by mean

df_pop = df.groupby(['country']).agg({'suicides_per_100k_pop': np.mean}).sort_values(by='suicides_per_100k_pop', ascending=False)
df_pop
```

```python
# merge with left join

df_final = df_gdp.merge(df_pop, on='country', how='left')
print(df_final.head())
```

```python
df_final.info()
```

```python
# Drop the rows where at least one element is missing

df_final.dropna(inplace=True)
df_final
```

```python
df_final.info()
```

```python
# Sorting DataFrame on a Single Column in descending order

top_10_final = df_final.sort_values(by="suicides_per_100k_pop", ascending=False)[:10]
top_10_final
```

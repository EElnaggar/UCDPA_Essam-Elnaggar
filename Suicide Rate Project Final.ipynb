```python
# Import the GDP table by country for the last year

table_gdp = pd.read_html('https://www.macrotrends.net/countries/ranking/gdp-per-capita')
df_gdp = table_gdp[0]
df_gdp.head(5)
```

```python
df_gdp.drop(['2018', 'currency'], inplace=True, axis = 1)
```

```python
df_gdp.head(5)
```

```python
# Removing a comma from a column values in Pandas DataFrame

df_gdp['gdp_per_capita'] = df_gdp['gdp_per_capita'].str.replace(',', '')
```

```python
# convert "gdp_2018" from object to int64

df_gdp = df_gdp.astype({'gdp_per_capita':'int'})
```

```python
df_gdp.info()
```

```python
# create a list of our conditions
conditions = [
    (df_gdp['gdp_per_capita'] <= 1026),
    (df_gdp['gdp_per_capita'] > 1026) & (df_gdp['gdp_per_capita'] <= 3995),
    (df_gdp['gdp_per_capita'] > 3995) & (df_gdp['gdp_per_capita'] <= 12375),
    (df_gdp['gdp_per_capita'] > 12375)
    ]

# create a list of the values we want to assign for each condition
values = ['Low-income', 'Lower-middle-income', 'Upper-middle-income', 'High-income']

# create a new column and use np.select to assign values to it using our lists as arguments
df_gdp['income_group'] = np.select(conditions, values)

# display updated DataFrame
df_gdp.head()
```

```python
# create a function to classify countries based on suicide rate per 100k of population

def suicides_rate(value):
    if value < 15: 
        return "low" 
    else: 
        return "high"
    return "non"
    
df['suicides_per_100k_pop'] = df['suicides_per_100k_pop'].map(suicides_rate)
display(df.head())
```
